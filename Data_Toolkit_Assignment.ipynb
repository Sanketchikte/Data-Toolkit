**3. Explain the following terms:**



*   The difference in np.array, np.asarray and np.asanyarray.
*   The difference between Deep copy and shallow copy

**Ans.**

**The difference in np.array, np.asarray, and np.asanyarray**



*   **np.array:** This function creates a new array from the input data. By default, it makes a copy of the input data, meaning changes to the original data do not affect the new array and vice versa.
*   **np.asarray:** This function converts the input data to an array, but does not make a copy if the input is already an array. If the input is not an array, it will convert it to one.




*   **np.asanyarray:** Similar to np.asarray, but it will convert the input to an array only if it is not already an instance of ndarray or a subclass of ndarray. This function does not make a copy if the input is already an array or a subclass of an array.


**The difference between Deep copy and shallow copy**



*   **Deep Copy:** A deep copy creates a new object and recursively copies all objects found in the original object. This means changes to the elements in the deep copy do not affect the original object and vice versa.
*   **Shallow Copy:** A shallow copy creates a new object but does not create copies of the objects that the original object references. Instead, it only copies the references of the objects. So, changes to the elements in the shallow copy will reflect in the original object and vice versa.

**5. Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array
perform the following operations:**

a)Extract all even integers from array.

b)Extract all odd integers from arrayX

**Ans.**


```python
import numpy as np

# Create a NumPy array with random integers between 1 and 10 of shape (5, 6)
array = np.random.randint(1, 11, size=(5, 6))
print("Original Array:\n", array)

# Extract all even integers from the array
even_integers = array[array % 2 == 0]
print("\nEven Integers:\n", even_integers)

# Extract all odd integers from the array
odd_integers = array[array % 2 != 0]
print("\nOdd Integers:\n", odd_integers)
```

**14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib:**

a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a time-series dataset
date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')
temperature = np.random.uniform(low=15, high=30, size=100)
humidity = np.random.uniform(low=40, high=90, size=100)

df = pd.DataFrame({'Date': date_range, 'Temperature': temperature, 'Humidity': humidity})

# Set the Date column as the index
df.set_index('Date', inplace=True)

# Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color=color)
ax1.plot(df.index, df['Temperature'], color=color, label='Temperature')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # Instantiate a second y-axis that shares the same x-axis
color = 'tab:blue'
ax2.set_ylabel('Humidity', color=color)
ax2.plot(df.index, df['Humidity'], color=color, label='Humidity')
ax2.tick_params(axis='y', labelcolor=color)

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Display the plot
fig.tight_layout()  # Adjust the layout to make room for the labels
plt.show()
```

**15.  Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
tasks using Matplotlib:**

a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function (PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a NumPy array containing 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# a) Plot a histogram of the data with 30 bins
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# b) Overlay a line plot representing the normal distribution's probability density function (PDF)
# Calculate the PDF
pdf = norm.pdf(bins, loc=0, scale=1)
plt.plot(bins, pdf, 'k--', linewidth=2)

# c) Label the x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Display the plot
plt.show()
```

**17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate two random arrays
x = np.random.randn(100)
y = np.random.randn(100)

# Determine the quadrant of each point
quadrants = ['Q1' if (xi >= 0 and yi >= 0) else
             'Q2' if (xi < 0 and yi >= 0) else
             'Q3' if (xi < 0 and yi < 0) else
             'Q4' for xi, yi in zip(x, y)]

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrants})

# Create a scatter plot
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(x='x', y='y', hue='Quadrant', palette='Set1', data=df)

# Add legend
scatter_plot.legend(title='Quadrant')

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title
plt.title('Quadrant-wise Scatter Plot')

# Show the plot
plt.show()
```

**18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
Wave Function'.**

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Enable the output to be displayed in the notebook
output_notebook()

# Generate the data for the sine wave
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Create a new plot
p = figure(title='Sine Wave Function', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label='Sine Wave', line_width=2)

# Add grid lines
p.xgrid.visible = True
p.ygrid.visible = True

# Show the results
show(p)
```

**19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'.**

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool

# Enable the output to be displayed in the notebook
output_notebook()

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=len(categories))

# Create a DataFrame
df = pd.DataFrame({'Category': categories, 'Value': values})

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create a bar chart
p = figure(x_range=categories, title='Random Categorical Bar Chart', x_axis_label='Category', y_axis_label='Value')

# Add bars with colors based on their values
p.vbar(x='Category', top='Value', width=0.9, source=source, legend_field='Category',
       fill_color='blue', line_color='black')

# Add hover tooltips to display exact values
hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

# Set the title of the plot
p.title.text = 'Random Categorical Bar Chart'

# Show the results
show(p)
```

**20.  Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as

**21. Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set
the title as 'Interactive Pie Chart'.**

```python
import numpy as np
import plotly.graph_objects as go

# Generate random data
labels = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 50, size=len(labels))

# Create a pie chart
fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=0.3)])

# Add labels and percentages
fig.update_traces(textinfo='label+percent', insidetextorientation='radial')

# Set the title
fig.update_layout(title_text='Interactive Pie Chart')

# Display the plot
fig.show()
```
